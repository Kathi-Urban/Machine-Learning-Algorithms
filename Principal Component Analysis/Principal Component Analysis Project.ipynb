## Principal Component Analysis (PCA) - Handwritten Digits

### Import and Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
digits = pd.read_csv('../Principal Component Analysis/digits.csv')
```

```python
digits
```

```python
# new DataFrame called pixels that consists only of the pixel feature values by dropping the number_label column

pixels = digits.drop('number_label', axis=1)
pixels
```

### Displaying an Image

```python
# single image row representation by getting the first row of the pixels DataFrame

single_image = pixels.iloc[0]
single_image
```

```python
# single row Series into a numpy array

single_image.to_numpy()
```

```python
single_image.to_numpy().shape
```

```python
# Reshape this numpy array into an (8,8) array

single_image.to_numpy().reshape(8,8)
```

```python
# diplay array as an image representation of the number drawn

plt.imshow(single_image.to_numpy().reshape(8,8))
```

```python
plt.imshow(single_image.to_numpy().reshape(8,8),cmap='gray')
```

```python
sns.heatmap(single_image.to_numpy().reshape(8,8),annot=True,cmap='gray')
```

### Scaling Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_pixels = scaler.fit_transform(pixels)
```

```python
scaled_pixels
```

### PCA

```python
from sklearn.decomposition import PCA
```

```python
pca_model = PCA(n_components=2)
```

```python
pca_pixels = pca_model.fit_transform(scaled_pixels)
```

```python
# variance explained by 2 principal components

np.sum(pca_model.explained_variance_ratio_)
```

```python
# scatterplot of the digits in the 2 dimensional PCA space

plt.figure(figsize=(10,6),dpi=150)
labels = digits['number_label'].values
sns.scatterplot(x = pca_pixels[:,0],y = pca_pixels[:,1],hue=labels,palette='Set1')
plt.legend(loc=(1.05,0))
```

Which numbers are the most "distinct"?

Label #4 being the most separated group, implying its the most distinct, similar situation for #2, #6 and #9.
